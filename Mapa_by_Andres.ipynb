```python
import leafmap
import pandas as pd

# Cargar tus datos
df = pd.read_csv("datos_consolidados_20251104_141743.csv")

# Asegurar tipos correctos
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.dropna(subset=['latitud', 'longitud'])

# Crear mapa base centrado en el promedio de tus estaciones
m = leafmap.Map(center=[df['latitud'].mean(), df['longitud'].mean()], zoom=8)

m.add_basemap("Esri.WorldImagery")  # Mapa satelital
# Tambin puedes probar: "OpenStreetMap", "HYBRID", "ROADMAP"

```

```python
import geopandas as gpd
from shapely.geometry import Point

# Convertir a GeoDataFrame
gdf = gpd.GeoDataFrame(
    df, geometry=[Point(xy) for xy in zip(df.longitud, df.latitud)], crs="EPSG:4326"
)

# Aadir capa de puntos al mapa
m.add_points_from_xy(
    df,
    x="longitud",
    y="latitud",
    color_column="pm_2p5_media_ugm3",
    layer_name="PM2.5",
    add_legend=True,
    popup=["nombre_estacion", "timestamp", "pm_2p5_media_ugm3"],
)
m
```
